**A Spring-Mass Model**

-Sometimes it is necessary to consider the second derivative when modeling a phenomenon.

We can construct a differential equation that models our oscillating mass. First, we must consider the restorative force on the spring. We will make the assumption that this force depends on the displacement of the spring, F(x). Using Taylor’s Theorem from calculus,we can expand F to obtain.



$F(x)=F(0)+F^\prime(0)x+\frac{1}{2}F^{\prime\prime}(0)x^2+...
=-Kx+\frac{1}{2}F^{\prime\prime}(0)x^2$


where F′(0) = −k and F(0) = 0. If the displacement is not too large, then xn will be small for n ≥ 2, and we can ignore higher ordered terms.



 Thus, we can consider the restorative force on the spring to be proportional to displacement of the spring from its equilibrium length.



$Hooke's Law:  F=-Kx$


This equation is known as Hooke’s Law. We can test this law experimentally, and it is reasonably accurate if the displacement of the spring is not too large.

By Newton’s second law of motion, the force on the mass must be


$ F = ma = m\frac{d^2x}{dt^2} = mx^{\prime\prime}$
.

Setting the two forces equal, we have a second-order differential equation.


$ mx^{\prime\prime} = −kx$



which describes our oscillating mass. The spring-mass system is an example of a harmonic oscillator. Harmonic oscillators are useful for modeling simple harmonic motion in mechanics.
