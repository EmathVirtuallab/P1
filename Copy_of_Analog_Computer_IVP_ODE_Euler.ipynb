```python
def matprint(mat, fmt="g"):
    col_maxes = [max([len(("{:"+fmt+"}").format(x)) for x in col]) for col in mat.T]
    for x in mat:
        for i, y in enumerate(x):
            print(("{:"+str(col_maxes[i])+fmt+"}").format(y), end="  ")
        print("")
```

# Matrix Version Predator-Prey System

# Matrix Lorenz System

$\frac{dx}{dt}=\sigma(y-x)$

$\frac{dy}{dt}=x(p-z)-y$

$\frac{dz}{dt}=xy-\beta z$

```python
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0
```

```python
model = np.array([[-sigma,sigma,0,1,0],[rho,-1,0,0,-1],[0,0,-beta,1,0]])
```

```python
print (model)
```

```python
plt.matshow(model)
```

```python
def dx_dt(x, t):
    x, y, z = x
    state = np.array([x,y,z,x*y,x*z])
    return model @ state 
```

```python
d = 2
x = np.zeros(d,)

x[0] = 10
x[1] = 10

t  = 1
tf = 50
dt = 0.001

n = int((tf-t)/dt)
X = np.zeros((n,d))

plt.plot(X);
plt.grid()
```

```python
for i in range(n):
x = x + dx_dt(x,t)*dt;
X[i] = x
plt.plot(x)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(X[:, 0], X[:, 1], X[:, 1]);
```
